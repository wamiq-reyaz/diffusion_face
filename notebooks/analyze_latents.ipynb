```python
import pickle
import torch
import os
import sys
import numpy as np
import tqdm
```

```python
raw_data_path = '../data/generated_samples/w_plus/all_wpluss.pkl'

with open(raw_data_path, 'rb') as fd:
    raw_data = pickle.load(fd)
```

```python

all_keys = list(raw_data.keys())
exemplar = raw_data[all_keys[0]]
data_keys = [k[0] for k in exemplar]

print(data_keys)
print(len(data_keys))

ALL_DATA = torch.empty(100000, len(data_keys), 512)
LENS = []

for ii, seed in tqdm.tqdm(enumerate(all_keys), total=len(all_keys)):
    curr_ws = raw_data[seed]
    for jj, (k, v) in enumerate(curr_ws):
        v = v.squeeze()
        valid_len = len(v)
        ALL_DATA[ii, jj, :valid_len] = v



ALL_DATA[0]
```

```python
_min = torch.min(ALL_DATA, dim=0)
_max = torch.max(ALL_DATA, dim=0)

print(_min.shape, _max.shape)
```

```python
import torch
import pickle
with open('../data/generated_samples/w_plus_img/all_wpluss_75000.pkl', 'rb') as fd:
    img_data = pickle.load(fd)

```

```python
new_keys = list(img_data.keys())
```

```python
print(img_data[new_keys[0]][-1][-1].dtype)
import matplotlib.pyplot as plt

plt.imshow(img_data[new_keys[0]][-1][-1].squeeze().numpy())
```

```python
ALL_IMG_DATA = torch.empty(75000, 3, 512, 512, dtype=torch.uint8)

for ii, k in enumerate(new_keys):
    curr_img = img_data[k][-1][-1].squeeze()
    curr_img = curr_img.permute(2, 0, 1)
    ALL_IMG_DATA[ii, ...] = curr_img
```

```python
# save images as PIL pngs
from PIL import Image
import tqdm

os.makedirs('../data/generated_samples/w_plus_img/images/', exist_ok=True)
for ii, curr_tensor in tqdm.tqdm(enumerate(ALL_IMG_DATA), total=len(ALL_IMG_DATA)):
    curr_tensor = curr_tensor.cpu()
    curr_img = curr_tensor.permute(1,2,0).numpy()
    Image.fromarray(curr_img).save(f'../data/generated_samples/w_plus_img/images/{str(ii).zfill(7)}.png')
    
```
